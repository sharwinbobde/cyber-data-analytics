```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
means ={}
std = {} 
```

```python
x = pd.read_csv('data/bga_k_times.txt', header=None)
means['bga_k'] = np.mean(x[0])
std['bga_k'] = np.std(x[0])
```

```python
x = pd.read_csv('data/rfgsm_k_times.txt', header=None)
means['rfgsm_k'] = np.mean(x[0])
std['rfgsm_k'] = np.std(x[0])
```

```python
x = pd.read_csv('data/topk_times.txt', header=None)
means['top_k'] = np.mean(x[0])
std['top_k'] = np.std(x[0])
```

```python
x = pd.read_csv('data/topk_plus_times.txt', header=None)
means['top_k +'] = np.mean(x[0])
std['top_k +'] = np.std(x[0])
```

```python
print(means)
print(std)
```

```python
methods = ['bga_k', 'rfgsm_k', 'top_k', 'top_k +']
x_pos = np.arange(len(methods))
heights = []
errors = []
for m in methods:
    heights.append(means[m])
    errors.append(std[m])
```

```python
fig, ax = plt.subplots()
fig.set_size_inches((10,6))
ax.bar(x_pos, heights, yerr=errors, align='center', alpha=0.6, ecolor='black', capsize=10)
ax.set_ylabel('Runtime in sec')
ax.set_xticks(x_pos)
ax.set_xticklabels(methods)
ax.set_title('Mean runtime for denfence along with standard deviation for 1 epoch')
ax.yaxis.grid(True)
```
